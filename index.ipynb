## Step 1: Read the dataset and inspect its columns and 5-point statistics

```python
# Load necessary libraries and import the data
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from statsmodels.formula.api import ols
%matplotlib inline

advertising=pd.read_csv("advertising.csv")
```

```python
# Check the columns and first few rows
advertising.head()
```

```python
# Get the 5-point statistics for data 
advertising.describe()
```

```python
advertising.drop(["Unnamed: 0"],axis=1,inplace=True)
advertising.head()
```

### Describe the contents of this dataset
It looks like it's got amounts of money spent in three categories: TV, radio, and newspaper as well as how many sales were made.

TV advertising has the highest mean amount while radio has the lowest.  The category with the most variation of money spent was newspaper advertising (because its standard deviation is a higher percentage of its mean as compared with the other advertising categories).

## Step 2: Plot histograms with kde overlay to check the distribution of the predictors

```python
## For all the variables, check distribution by creating a histogram with kde
for column in advertising:
    sns.distplot(advertising[column])
    plt.show()
```

### Record your observations here 
It looks like TV looks uniform-ish (or maybe left skew?) as a bar graph but bimodal as a KDE. Radio is a reverse J-shaped. Newspaper and sales are both skew right.


## Step 3: Test for the linearity assumption 

Use scatterplots to plot each predictor against the target variable

```python
# visualize the relationship between the preditors and the target using scatterplots
for column in advertising:
    if column!="sales":
        sns.scatterplot(advertising[column],advertising["sales"],label=column)
        plt.show()
    else:
        continue
```

### Record your observations on linearity here 
It looks like TV might be a square root or logistic.  Radio could be linear and there doesn't seem to be any relationship between newspaper advertising and sales.

### Conclusion so far

Based on above initial checks, we can confidently say that TV and radio appear to be good predictors for our regression analysis. Newspaper is very heavily skewed and also doesnt show any clear linear relationship with the target. 
> We'll move ahead with our analysis using TV and radio, and rule out newspaper because we believe it violates OLS assumptions

Note: Kurtosis can be dealt with using techniques like log normalization to "push" the peak towards the center of distribution. You'll learn about this later on.

## Step 4: Run a simple regression in Statsmodels with TV as a predictor

```python
# build the formula 
f = 'sales~TV'

# create a fitted model in one line
model = ols(formula=f, data=advertising).fit()
```

## Step 5: Get Regression Diagnostics Summary

```python
model.summary()
```

Note here that the coefficients represent associations, not causations

## Step 6:  Draw a prediction line with data points on a scatter plot for X (TV) and Y (Sales)

Hint: You can use the `model.predict()` function to predict the start and end point of of regression line for the minimum and maximum values in the 'TV' variable. 

```python
# create a DataFrame with the minimum and maximum values of TV
df_tv=pd.DataFrame({'TV':[advertising["TV"].min(),advertising["TV"].max()]})
print(df_tv.head())
# make predictions for those x values and store them

predictions=model.predict(df_tv['TV'])
print("-----------")
print("Predictions")
print(predictions)
# first, plot the observed data and the least squares line
sns.scatterplot(advertising["TV"],advertising["sales"],color="blue",label="data")
sns.lineplot(df_tv["TV"],predictions,color="red",label="regression")
```

## Step 7: Visualize the error term for variance and heteroscedasticity

```python
fig = plt.figure(figsize=(16,8))
fig = sm.graphics.plot_regress_exog(model, "TV", fig=fig)
plt.show()
```

### Record Your observations on heteroscedasticity
It looks like there's heteroscedasticity when we look at the top graph.  The residuals aren't uniformly/evenly spaced from the regression line.  And, if it shows hetero- instead of homoscedasticity, then we can't use OLS as a model.

## Step 8: Check the normality assumptions by creating a QQ-plot

```python
# Code for QQ-plot here
import scipy.stats as stats
residuals = model.resid
fig = sm.graphics.qqplot(residuals, dist=stats.norm, line='45', fit=True)
fig.show()
```

### Record Your observations on the normality assumption
The QQ plot shows a pretty straight line so it's probably safe to assume that the residuals are normally distributed.

## Step 9: Repeat the above for radio and record your observations

```python
# code for model, prediction line plot, heteroscedasticity check and QQ normality check here
f = 'sales~radio'
model = ols(formula=f, data=advertising).fit()
```

```python
model.summary()
```

```python
# create a DataFrame with the minimum and maximum values of TV
df_radio=pd.DataFrame({'radio':[advertising["radio"].min(),advertising["radio"].max()]})
print(df_tv.head())
# make predictions for those x values and store them

predictions=model.predict(df_radio['radio'])
print("-----------")
print("Predictions")
print(predictions)
# first, plot the observed data and the least squares line
sns.scatterplot(advertising["radio"],advertising["sales"],color="blue",label="data")
sns.lineplot(df_radio["radio"],predictions,color="red",label="regression")
```

```python
fig = plt.figure(figsize=(16,8))
fig = sm.graphics.plot_regress_exog(model, "radio", fig=fig)
plt.show()
```

```python
residuals = model.resid
fig = sm.graphics.qqplot(residuals, dist=stats.norm, line='45', fit=True)
fig.show()
```

### Record your observations here for goodnes of fit 
Radio also shows heteroscedasticity so it doesn't meet the requirements for using the OLS model.  The residuals may also not be normally distributed, since the QQ plot has a bit of a curve on it.
